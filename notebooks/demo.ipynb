# Simulate closed-loop
After implementing your control functionality, you can simulate the closed-loop with code that looks something like this...

```python
import sys
import os

# Use the correct relative path to the project root directory
sys.path.append(os.path.abspath('../'))

# Import the necessary classes
from uuv_mission.dynamic import Submarine, ClosedLoop, Mission
from uuv_mission.control import PDController

# Initialize the components
sub = Submarine()
controller = PDController(Kp=0.1, Kd=0.7)  # PD controller with given gains
closed_loop = ClosedLoop(sub, controller)

# Load mission data
mission = Mission.from_csv("../data/mission.csv")

# Simulate the closed-loop system with random disturbances
trajectory = closed_loop.simulate_with_random_disturbances(mission)

# Plot the completed mission
trajectory.plot_completed_mission(mission)
```
